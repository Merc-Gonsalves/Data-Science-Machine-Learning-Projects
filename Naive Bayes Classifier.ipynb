```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris, load_digits
```

```python
df = pd.read_csv("/mnt/c/Users/mercg/Downloads/titanic.csv")
df.head()
```

```python
df.drop(columns=['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], inplace=True)
df.head()
```

```python
X = df.drop(columns='Survived')
y = df.Survived
```

```python
dummies = pd.get_dummies(X['Sex'])
X = pd.concat([X, dummies], axis= 'columns')
X
```

```python
X['Age'] = X['Age'].fillna(X.Age.mean())
X.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X.drop(columns='Sex', inplace=True)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
nb.score(X_test, y_test)
```

```python
cross_val_score(nb, X_train, y_train).mean()
```

```python
df1 = pd.read_csv("/mnt/c/Users/mercg/Downloads/spam.csv")
df1.head()
```

```python
df1['Category'].unique()
```

```python
df1['spam'] = df1['Category'].map({'spam': 1, 'ham': 0})
df1
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1.Message, df1.spam, test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_train_count = cv.fit_transform(X_train.values)
X_test_count = cv.transform(X_test.values)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
model.score(X_test_count, y_test)
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([
    ('vec', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```
