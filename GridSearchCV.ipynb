```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold, cross_val_score
from sklearn.datasets import load_iris, load_digits
from sklearn.model_selection import GridSearchCV
```

```python
iris = load_iris()
```

```python
tuning = GridSearchCV(SVC(gamma = 'auto'), {
    'C':[1,10,20],
    'kernel':['linear', 'poly', 'rbf']
}, cv = 5, return_train_score=False)

tuning.fit(iris.data, iris.target)
tuning.cv_results_
```

```python
df = pd.DataFrame(tuning.cv_results_)
df
```

```python
tune = GridSearchCV(RandomForestClassifier(),{
    'n_estimators': [10,30,50,80],
    'criterion': ['gini', 'entropy', 'log_loss']
})
```

```python
tune.fit(iris.data, iris.target)
tune.best_score_
```

```python
res = pd.DataFrame(tune.cv_results_)
res
```
